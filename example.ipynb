```python
import seaborn as sns
import numpy as np
from stat_utils import SpatialDistribution
```

## Same distribution

```python
a = np.random.normal(1, 0.5, (1000, 2))
b = np.random.normal(1, 0.5, (1000, 2))
sns.jointplot(x=np.hstack([a[:, 1], b[:, 1]]), y=np.hstack([a[:, 0], b[:, 0]]), kind="kde")
```

```python
sd = SpatialDistribution()
sd.fit(a, b)
sd.simulate(n_iter=1000)
sd.summary()
```
